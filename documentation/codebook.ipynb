```python
import pandas as pd

df = pd.read_csv('/Users/jasmineliu/Downloads/QTM350/Final project/dataset/employment.csv')

codebook_employment_data = pd.DataFrame({
    'Variable Name': df.columns,
    'Data Type': [df[col].dtype for col in df.columns],
    'Non-Null Count': df.notnull().sum().values,
    'Unique Values': [df[col].nunique() for col in df.columns],
    'Sample Value': [df[col].dropna().iloc[0] if not df[col].dropna().empty else 'N/A' for col in df.columns]
})

codebook_employment_data.to_csv('codebook_employment_data.csv', index=False)
```
